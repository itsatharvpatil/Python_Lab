```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn import metrics
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
heart_risk = pd.read_csv("C:/Users/Atharv/Downloads/Session_3_-_Regression/logreg_train_test.csv")
```

```python
heart_risk.isnull().sum()
```

```python
heart_risk.drop(columns = ['education'], inplace = True)
```

```python
heart_risk['cigsPerDay'] = heart_risk['cigsPerDay'].fillna(heart_risk['cigsPerDay'].mean())
heart_risk['BPMeds'] = heart_risk['BPMeds'].fillna(heart_risk['BPMeds'].mean())
heart_risk['totChol'] = heart_risk['totChol'].fillna(heart_risk['totChol'].mean())
heart_risk['BMI'] = heart_risk['BMI'].fillna(heart_risk['BMI'].mean())
heart_risk['glucose'] = heart_risk['glucose'].fillna(heart_risk['glucose'].mean())
heart_risk['heartRate'] = heart_risk['heartRate'].fillna(heart_risk['heartRate'].mean())
```

```python
heart_risk.describe().transpose()
```

```python
heart_risk.groupby(["TenYearCHD"]).count()
```

```python
#sns.pairplot(heart_risk,diag_kind = 'kde',hue = 'TenYearCHD')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = heart_risk.drop('TenYearCHD', axis=1)
Y = heart_risk['TenYearCHD']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.1,random_state = 42)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
heart_risk.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='lbfgs',max_iter=10000)
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
model.score = model.score(X_test,y_test)

print(model.score)
print(metrics.confusion_matrix(y_test,y_predict))
```
